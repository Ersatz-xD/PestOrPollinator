# Pest or Pollinator: Exploratory Data Analysis (EDA)

Before we train any models, we need to understand the data we are working with. This notebook performs a baseline EDA on our 100-species dataset.

**Our goals for this phase are to:**
1. **Check for Class Imbalance:** Ensure our model won't become biased towards overrepresented species (the "lazy student" problem).
2. **Profile the Species:** Cross-reference our 100 classes to begin identifying which are agricultural pests and which are beneficial pollinators.

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("gpiosenka/butterfly-images40-species")

print("Path to dataset files:", path)
```

```python
import pandas as pd
from pathlib import Path
```

## 1. Traversing the Folders & Checking for Imbalance

Since this is an image dataset, our labels are defined by the folder structure (the `ImageFolder` format). We will iterate through the `train` directory, count the images inside each species folder, and calculate the distribution to see if we have a healthy dataset.

```python
dataset_path = Path(path) / 'train'
class_counts = {}

for class_dir in dataset_path.iterdir():
    if class_dir.is_dir():
        file_count = len(list(class_dir.glob('*')))
        class_counts[class_dir.name] = file_count

df_counts = pd.DataFrame(list(class_counts.items()), columns=['Species', 'ImageCount'])

print("DATASET IMBALANCE CHECK")
print(f"Total Species Classes: {len(df_counts)}")
print(f"Average Images per Species: {df_counts['ImageCount'].mean():.2f}")
print(f"Max Images (Most Common): {df_counts['ImageCount'].max()} ({df_counts.loc[df_counts['ImageCount'].idxmax(), 'Species']})")
print(f"Min Images (Rarest): {df_counts['ImageCount'].min()} ({df_counts.loc[df_counts['ImageCount'].idxmin(), 'Species']})")
print("-" * 30)
```
